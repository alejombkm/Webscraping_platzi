```python
from bs4 import BeautifulSoup
```

```python
s = BeautifulSoup(p12.text, "html")
```

```python
type(s)
```

```python
print(s.prettify())
```

```python
sections = s.find('ul', attrs={'class':'horizontal-list main-sections hide-on-dropdown'}).find_all('li')
sections
```

```python
section = sections[0]
```

```python
section.a
```

```python
section.find('a')
```

```python
section.a.get('href')
```

```python
section.a.get_text()
```

```python
link_sections = [section.a.get('href') for section in sections]
link_sections
```

```python
sec = requests.get(link_sections[0])
```

```python
sec.status_code
```

```python
s_section = BeautifulSoup(sec.text,'lxml')
```

```python
print(s_section.prettify())
```

```python
featured_article = s_section.find('section', attrs={'class':'top-content'} )
featured_article
```

```python
#get URL
featured_article.a.get('href')
```
