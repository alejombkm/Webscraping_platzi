```python
def get_links(soup):
    '''
    funtion to recive a object by BeautifulSoup and return a list of URL
    '''
    list_notes = []

    #get promoted article
    featured_article = soup.find('div', attrs={'class':'top-content'})
    if featured_article:
        list_notes.append(featured_article.a.get('href'))

    #get list articles
    list_articles = soup.find('div', attrs={'class':'articles-list'})
    for article in list_articles.find_all():
        if article.a:
            list_notes.append(article.a.get('href'))

    return list_notes
```

# Downloading content

```python
list_notes = get_links(s_section)
url_nota = url + list_notes[0]
url_nota
```

```python
try:
    note = requests.get(url_nota)
    if note.status_code == 200:
        s_note = BeautifulSoup(note.text, 'lxml')
        #Extract title
        title = s_note.find('h1')
        #Extract date
        date = s_note.find('div', attrs = {'class':'date modification-date'}).span.time.get('datetime')
        #Extract Volanta
        volanta = s_note.find('h2')
        #Extract firs paragraph
        firt_p = s_note.find('h2', attrs = {'class':'h3'})
        print(" The title is: \n",title.text ,"\n Date of article is: \n", date,"\n volanta: \n",volanta.text, "\n first paragraph: \n ",firt_p.text)
except Exception as e:
    print("Error: \n", e, "\n")
```
