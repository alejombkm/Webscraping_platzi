```python
def get_links(soup):
    '''
    funtion to recive a object by BeautifulSoup and return a list of URL
    '''
    list_notes = []

    #get promoted article
    featured_article = soup.find('div', attrs={'class':'top-content'})
    if featured_article:
        list_notes.append(featured_article.a.get('href'))

    #get list articles
    list_articles = soup.find('div', attrs={'class':'articles-list'})
    for article in list_articles.find_all():
        if article.a:
            list_notes.append(article.a.get('href'))

    return list_notes
```

```python
def obtener_info(s_nota):
    
    # Creamos un diccionario vacío para poblarlo con la información
    ret_dict = {}
    
    # Extraemos la fecha
    fecha = s_note.find('div', attrs = {'class':'date modification-date'}).span.time.get('datetime')
    if fecha:
        ret_dict['fecha'] = fecha
    else:
        ret_dict['fecha'] = None
    
    # Extraemos el título
    titulo = s_nota.find('div', attrs={'class':'article-title'})
    if titulo:
        ret_dict['titulo'] = titulo.text
    else:
        ret_dict['titulo'] = None

    # Extraemos la volanta
    volanta = s_nota.find('div', attrs={'class':'article-prefix'})
    if volanta:
        ret_dict['volanta'] = volanta.get_text()
    else:
        ret_dict['volanta'] = None
    
    # Extraemos el copete
    copete = s_nota.find('div', attrs={'class':'article-summary'})
    if copete:
        ret_dict['copete'] = volanta.get_text()
    else:
        ret_dict['copete'] = None
    
    autor = s_nota.find('div', attrs={'class':'article-author'})
    if autor:
        ret_dict['autor'] = autor.a.get_text()
    else:
        ret_dict['autor'] = None
    
    # Extraemos la imagen
    media = s_nota.find('div', attrs={'class':'article-main-media-image'})
    if media:
        imagenes = media.find_all('img')
        if len(imagenes) == 0:
            print('no se encontraron imágenes')
        else:
            imagen = imagenes[-1]
            img_src = imagen.get('data-src')
            try:
                img_req = requests.get(img_src)
                if img_req.status_code == 200:
                    ret_dict['imagen'] = img_req.content
                else:
                    ret_dict['imagen'] = None
            except:
                print('No se pudo obtener la imagen')
    else:
        print('No se encontró media')
    
    # Extraemos el cuerpo de la nota
    cuerpo = s_nota.find('div', attrs={'class':'article-text'})
    if cuerpo:
        ret_dict['texto'] = cuerpo.get_text()
    else:
        ret_dict['texto'] = None
    
    return ret_dict
```

# Unifying the scraper

```python
def scrape_note(url):
    try:
        note = requests.get(url)
    except Exception as e:
        print('No found URL \n', url, "\n error \n", e)
        return None

    if note.status_code != 200:
        print(f"Error get note {url} ,status code{note.status_code}")
        return None

    s_note = BeautifulSoup(note.text, 'lxml')
    
    ret_dict = obtener_info(s_note)
    ret_dict['url']= url

    return ret_dict
```
